# **Inference Result Comparison**

```python
import os
import re
import pandas as pd
from IPython.display import display, Markdown

pd.set_option("display.max_colwidth", None)
```

```python
base_dir = "."
scenario_dirs = [
    d for d in os.listdir(base_dir)
    if os.path.isdir(d) and d not in ("__pycache__",)
]
```

```python
scenario_dirs
```

```python
scenario_dirs = ['base', 'sft20', 'sft45',
                 'sft_only', 'dpo70', 'dpo45', 'dpo_only']
```

```python
MODEL_ORDER = [
    "Meta-Llama-3.1-8B",
    "Aya-23-8B",
    "SeaLLMs-v3-7B",
    "SEA-LION-v3-8B",
    "Sahabat-AI-8B",
]
```

```python
# discover your scenarios
scenarios = {}
for d in scenario_dirs:
    models = {}
    for fn in os.listdir(d):
        if fn.endswith(".jsonl"):
            m = re.match(r".*detailed_(.+)\.jsonl$", fn)
            if m:
                models[m.group(1)] = os.path.join(d, fn)
    if models:
        scenarios[d] = models

# load, merge, reorder, and display
for scenario, model_files in scenarios.items():
    dfs = []
    for model_key, filepath in model_files.items():
        df = pd.read_json(filepath, lines=True)
        col = f"{scenario}_{model_key}"
        dfs.append(df.rename(columns={model_key: col})[["prompt","ground_truth", col]])
    # inner‐merge all on prompt + ground_truth
    merged = dfs[0]
    for df in dfs[1:]:
        merged = merged.merge(df, on=["prompt","ground_truth"], how="outer")

    # build column list in desired order
    cols = ["prompt", "ground_truth"]
    for m in MODEL_ORDER:
        col_name = f"{scenario}_{m}"
        if col_name in merged.columns:
            cols.append(col_name)

    display(Markdown(f"### Scenario: **{scenario}**"))
    # random number of rows to display
    display(merged[cols].tail(2))
```
